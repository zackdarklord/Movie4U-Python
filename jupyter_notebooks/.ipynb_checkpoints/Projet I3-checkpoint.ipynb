## Systeme de recommandation

Un systme de recommandation est un type de logiciel ou d'application qui utilise des techniques de traitement de donnes et d'apprentissage automatique pour suggrer des lments ou des contenus pertinents  un utilisateur. Ces lments peuvent inclure des produits, des services, des articles, des films, de la musique, des amis, des informations, etc., en fonction des besoins, des prfrences et du comportement de l'utilisateur. Le but principal d'un systme de recommandation est d'offrir une exprience personnalise  l'utilisateur, en lui proposant des choix qui correspondent  ses gots, ses intrts, ses habitudes et ses besoins.

L'une des approches les plus simples et nanmoins efficaces pour la recommandation de films en fonction des informations des utilisateurs est le filtrage collaboratif. Le filtrage collaboratif se divise en deux catgories principales :
* Filtrage collaboratif utilisateur-utilisateur : Cette approche recommande des films  un utilisateur en se basant sur les gots et les prfrences d'autres utilisateurs similaires. La similarit entre les utilisateurs peut tre calcule en utilisant des mesures telles que la similarit cosinus.
* Filtrage collaboratif article-article : Au lieu de se baser sur la similarit entre les utilisateurs, cette approche recommande des films en se basant sur la similarit entre les films eux-mmes. Si un utilisateur a aim un certain film, d'autres films similaires peuvent lui tre recommands.

# Nttoyage des donnes

## Nttoyage anne de sortie

```python
dfMaster[["title","Year"]] = dfMaster.Title.str.extract("(.)\s\((.\d+)",expand=True)
dfMaster = dfMaster.drop(columns=["title"])
dfMaster['Year'] = dfMaster.Year.astype(int)
dfMaster['Movie_Age'] = 2000 - dfMaster.Year
dfMaster.head()
```

## Nttoyage sur le genre utilisateur

```python
dfMaster['Gender'] = dfMaster.Gender.replace('F','1')
dfMaster['Gender'] = dfMaster.Gender.replace('M','0')
dfMaster['Gender'] = dfMaster.Gender.astype(int)
dfMaster.head()
```

## Nttoyage des noms de colonne

```python
col = []
for i in dfMaster.columns:
    i = i.replace('-','_')
    i = i.replace('(','_')
    i = i.replace(')','_')
    i = i.replace("'",'_')
    col.append(i)
dfMaster.columns = col
dfMaster.head()
```

# Prdiction pour la suggestion de films

## Construction du modle

```python
import json
# Fonction pour recommender en fonction d'un utilisateur qui dj des notes
def get_movie_recommendations(user_id, num_recommendations):
    
    # slection les films les mieux nots
    dfMaster_best = dfMaster[dfMaster["Rating"] >= 3].sort_values(by=['Rating','Year'], ascending=False)
    
    # Contruction du modle
    train_data, test_data = train_test_split(dfMaster_best, test_size=0.2)
    user_movie_matrix = train_data.pivot_table(index='UserID', columns='MovieID', values='Rating')
    user_movie_matrix = user_movie_matrix.fillna(0)
    user_similarity = cosine_similarity(user_movie_matrix)
    
    # vrification
    if user_id not in user_movie_matrix.index:
        print("L'utilisateur {} n'existe pas dans les donnes.".format(user_id))
        return "[]"

    # Slectionnez les films prfrs de l'utilisateur
    user_ratings = user_movie_matrix.loc[user_id]

    # Crez une liste de films dj nots par l'utilisateur
    rated_movies = user_ratings[user_ratings > 0].index

    # Crez une liste pour stocker les recommandations
    recommendations = []

    for movie_id in user_movie_matrix.columns:
        # Si l'utilisateur a dj not ce film, passez  l'itration suivante
        if movie_id in rated_movies:
            continue

        # Calculez la similarit entre l'utilisateur donn et les autres utilisateurs qui ont not ce film
        similarity = user_movie_matrix[user_movie_matrix[movie_id] > 0].dot(user_ratings)

        # Triez les utilisateurs similaires par ordre dcroissant de similarit
        sorted_users = similarity.sort_values(ascending=False)

        # Slectionnez le premier utilisateur (le plus similaire) et ajoutez le film  la liste des recommandations
        if not sorted_users.empty:
            most_similar_user = sorted_users.index[0]
            recommendations.append({"MovieID": movie_id, "Similarity": sorted_users[most_similar_user]})

        # Si nous avons atteint le nombre souhait de recommandations, sortez de la boucle
        if len(recommendations) >= num_recommendations:
            break

    # Triez les recommandations par score de similarit dcroissant
    recommendations = sorted(recommendations, key=lambda x: x["Similarity"], reverse=True)

    return recommendations
```

```python
def get_movie_info(movie_id):
    return dfMovies.iat[movie_id, 1] + ", Genre : " + dfMovies.iat[movie_id, 2]
    
def print_recommendation():
    # Utilisez la fonction pour obtenir les 10 meilleurs films recommands pour un utilisateur spcifique
    top_movie_recommendations = get_movie_recommendations(1, num_recommendations=10)
    print("Top 10 films recommands pour l'utilisateur 1 :")
    for i, recommendation in enumerate(top_movie_recommendations):
        print(f"{i+1}. ID: {recommendation['MovieID']}, Title: {get_movie_info(recommendation['MovieID'])}, Similarity: {recommendation['Similarity']}")

print_recommendation()
```
